```python
import datetime
import json
import pymysql
import pymongo
import schedule
import time
from pysnmp.smi import builder, view, error
from pysnmp.entity.rfc3413.oneliner import cmdgen

# 创建 cmd generator
cmdGen = cmdgen.AsynCommandGenerator()  # 異步

# 回调函数
# 最后一个参数是创建任务时传入的 callback context
def callback(sendRequestHandle, errorIndication, errorStatus, errorIndex, varBinds, cbCtx):
    print("ipA:", cbCtx["ip"])
    system = []
    ifTable = {}
    ifIndex = []
    ifDescr = []
    ifType = []
    ifMtu = []
    ifSpeed = []
    ifPhysAddress = []
    ifAdminStatus = []
    ifOperStatus = []
    ifLastChange = []
    ifInOctets = []
    ifInUcastPkts = []
    ifInNUcastPkts = []
    ifInDiscards = []
    ifInErrors = []
    ifInUnknownProtos = []
    ifOutOctets = []
    ifOutUcastPkts = []
    ifOutNUcastPkts = []
    ifOutDiscards = []
    ifOutErrors = []
    ifOutQLen = []
    ifSpecific = []

    hrSystem = []

    hrStorage = {}
    hrMemorySize = 0
    hrStorageIndex = []
    hrStorageType = []
    hrStorageDescr = []
    hrStorageAllocationUnits = []
    hrStorageSize = []
    hrStorageUsed = []
    hrStorageAllocationFailures = []

    hrSWRun = {}
    hrSWRunIndex = []
    hrSWRunName = []
    hrSWRunID = []
    hrSWRunPath = []
    hrSWRunParameters = []
    hrSWRunType = []
    hrSWRunStatus = []
    hrSWRunPerfCPU = []
    hrSWRunPerfMem = []
    hrSWInstalled = {}
    hrSWInstalledIndex = []
    hrSWInstalledName = []
    hrSWInstalledID = []
    hrSWInstalledType = []
    hrSWInstalledDate = []
    if errorIndication:  # SNMP agent errors
        print("errorIndication:", errorIndication)
    else:
        if errorStatus:
            print('%s at %s' % (errorStatus.prettyPrint(), varBinds[int(errorIndex)-1] if errorIndex else '?'))
        else:
            for row in varBinds:
                for oid, val in row:
                    if cbCtx["oid"] not in oid.prettyPrint():
                        return 0  # 告诉 walk 停止运行
                    snmpContext = {}
                    snmpContext["oid"] = oid.prettyPrint()
                    snmpContext["val"] = val.prettyPrint()
                    oidKey = getOID(oid.prettyPrint())
                    snmpContext["oid"] = oidKey
                    snmpContext["val"] = val.prettyPrint()
                    print("oid:", getOID(oid.prettyPrint()), "val:", val.prettyPrint())

            print("Data Done")
            return 1  # 告诉 walk 继续运行
    print("Done")

def getOID(strword):
    s1 = "."
    strword = strword.replace("SNMPv2-SMI::mib-2", "1.3.6.1.2.1")
    arr = strword.split(".")
    del arr[-1]
    return s1.join(arr)

# hosts = ["140.134.213.101", "140.134.211.50", "140.134.211.4"]
hosts = ["127.0.0.1"]
oids = ['1.3.6.1.2.1.1','1.3.6.1.2.1.2.2','1.3.6.1.2.1.25.1','1.3.6.1.2.1.25.2','1.3.6.1.2.1.25.4','1.3.6.1.2.1.25.5','1.3.6.1.2.1.25.6']
# oids = ['1.3.6.1.2.1.1']

nowDate = datetime.datetime.now()

for host in hosts:
#     print("ip:", host)
    snmpJson = {}
    snmpJson["receiveTime"] = nowDate
    for oid in oids:
        snmpJson["ip"] = host
        snmpJson["oid"] = oid
        print("oid:", oid)
        cmdGen.nextCmd(
            # snmp v2c
            cmdgen.CommunityData('public'),
            cmdgen.UdpTransportTarget((host, 161)),
            (cmdgen.MibVariable(oid),),
            (callback, snmpJson)
        )
#     print("snmpJson:", snmpJson)

# 啟動
cmdGen.snmpEngine.transportDispatcher.runDispatcher()
```
